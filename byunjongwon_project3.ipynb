```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
# Question 1: Use yfinance to Extract Stock Data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
# Question 2: Use Webscraping to Extract Tesla Revenue Data
!pip install selenium pandas beautifulsoup4 lxml html5lib

import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import StringIO

url_tesla = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data_tesla = requests.get(url_tesla).text
soup_tesla = BeautifulSoup(html_data_tesla, "html.parser")
tables_tesla = soup_tesla.find_all("table")

tesla_revenue = pd.read_html(StringIO(str(tables_tesla)))[0]
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue.dropna(inplace=True)

print("Tesla income data")
print(tesla_revenue.tail())
```

```python
# Question 3: Use yfinance to Extract Stock Data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
print(gme_data.head())
```

```python
# Question 4: Use Webscraping to Extract GME Revenue Data

import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import StringIO

url_gme = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_data_gme = requests.get(url_gme).text
soup_gme = BeautifulSoup(html_data_gme, "html.parser")
tables_gme = soup_gme.find_all("table")
gme_revenue = pd.read_html(StringIO(str(tables_gme)))[0]
gme_revenue.columns = ["Date", "Revenue"]
gme_revenue.dropna(inplace=True)

print("GameStop income data")
print(gme_revenue.tail())
```

```python
# Question 5: Plot Tesla Stock Graph
def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data["Date"], data["Close"], label=title)
    plt.xlabel("Date")
    plt.ylabel("Closing Price")
    plt.title(title)
    plt.legend()
    plt.show()

make_graph(tesla_data, "Tesla Stock Price History")
```

```python
# Question 6: Plot GameStop Stock Graph
make_graph(gme_data, "GameStop Stock Price History")
```
